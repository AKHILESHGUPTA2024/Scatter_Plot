```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
plt.scatter(x,y,s=x*300,c=np.where(y>.5,'g','r'),alpha=0.5,edgecolors='k')
```

```python
x=np.random.normal(5.0,1.0,500)
y=np.random.normal(10.0,2.0,500)
plt.style.use('fast')
plt.scatter(x,y,c=np.where(y>10,'b','r'),s=x*10,alpha=0.5,edgecolors='k',vmin=0,vmax=10,cmap='viridis',marker='o',linewidths=2)
```

```python
import matplotlib.cm as cm
x=np.arange(20)
ys = [i+x+(i*x)**2 for i in range(20)]
colors = cm.rainbow(np.linspace(0,1,len(ys)))
for y,c in zip(ys,colors):
    plt.scatter(x,y,color=c)
```

```python
import itertools
colors = itertools.cycle(['r','g','b'])
for y in ys:
    plt.scatter(x,y,color=next(colors),s=x**2)
```
